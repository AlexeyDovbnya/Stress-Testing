№1 "Информационные системы, подходящие под критерии НТ"

​1. Привести 3 примера информационных систем (сервисов) которые соответствуют критериям: - Высоконагруженная система, нагрузка создаётся либо реальными пользователями, либо другой информационной системой (например сервис хранения информации о штрафах гибдд). - Простой или неработоспособность данной системы дорого обходится владельцам или критичен для инфраструктуры, функционирование которой система обеспечивает система (сервис).

- МаркетПлейс - вэб сервис высоконагруженная система (реальные пользователи) сбой, простой, неработоспособность дорого обходится владельцам (товары не продаются) и критичен для инфраструктуры, которая обеспечивает система. Другая информационная система (товары не регестрируются/резервируются, не отправляются по назначению).
- Социальная сеть. Нагрузка создается как самими пользователями так и внутренними информационными сервисами (регистрационными, учетными и т.д.)
- Агрегаторы Рецензий (Кинопоиск и т.д.) Нагрузка создается как пользователями так и внутренними инфромационными системами (например системой рекомендаций).

2. Написать 3 пользовательских сценария для данной системы (всего 9 сценариев). Если нагрузка не пользовательская, написать 3 типа информации, которая поступает или запрашивается у системы. Привести примеры типичной модели нагрузки за определённый, также типичный отрезок времени для данной системы (год, месяц, день, час). На занятии уже озвучили: Банковские приложения, Маркет Плейсы, Стриминги.

Маркетплейс
  - Пользовательская нагрузка:
    - Вход, индентификация пользователя.
    - Ввод пользователем информации об интересующем товаре.
    - Ввод пользователем платежной информации.

    
  - Нагрузка на систему регистрации товаров (резервирования): магазин продавца, наличие/отсутствие товара, контактная информация.
  - Нагрузка на логистическую систему: время доставки/отгрузки, стоимость доставки, ответсвенный и т.д.





Социальная сеть
- Пользовательская нагрузка:
  - Вход, индентификация пользователя.
  - Загрузка пользователем медиа-файлов (фото, видео)
  - Пользователь написал и загрузил пост
- Внутренний платежный сервис (верификационная информация, id операции, id пользователя)
- Агрегирующая и рекомендательные системы сбора информации о пользователе (возраст, пол, друзья)

Агрегаторы Рецензий
  - Пользовательская нагрузка:
    - Вход, индентификация пользователя.
    - Сценарий поиска пользователем запрашиваемой информации исходя из его предпочтений.
    - Написание и загрузка постов, рецензий и проставленине оценок.
  - Система рекомендаций с использванием ИИ (информация о пользователе, оценки, время оценки)
  - система CV по распознаванию лиц актеров (подготовленные данные для нейросети - стоп-кадры фильмов, музыка для распознавания (скорее всего сторонний сервис через API, но нагрузка должна быть большая)
